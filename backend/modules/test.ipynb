```python
import pandas as pd
```

```python
df = pd.read_csv("D:\Scam_shield_PICT\dataset\BETTER30.csv")
```

```python
df.head()
```

```python
label_mapping = {
    "scam": "suspicious",
    "scam_response": "suspicious",
    "suspicious": "suspicious",
    "slightly_suspicious": "suspicious",
    "potential_scam": "suspicious",
    "highly_suspicious": "suspicious",
    "neutral": "not suspicious",
    "legitimate": "not suspicious",
}
```

```python
df["LABEL"] = df["LABEL"].str.strip().map(label_mapping).fillna("not suspicious")
```

```python
df = df[["TEXT", "LABEL"]]
```

```python
df["LABEL"] = df["LABEL"].apply(lambda x: 1 if x == "suspicious" else 0)
```

```python
df.to_csv("cleaned_scam_data.csv", index=False)
```
